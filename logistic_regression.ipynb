```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot  as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=10,random_state=42,n_classes=2,)
```

```python
df = pd.DataFrame(X)
df
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test  = train_test_split(X,y,test_size = 0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_reg = LogisticRegression()
```

```python
logistic_reg.fit(X_test)
```

```python
y_pred = logistic_reg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

acc = accuracy_score(y_test,y_pred)
acc
con = confusion_matrix(y_test,y_pred)
con

pre_recall =  classification_report(y_test,y_pred)

acc,con,
pre_recall
acc
```

# hyperparameter tuning 

```python
model = LogisticRegression()
```

```python
penalty = ['l1', 'l2', 'elasticnet']
c_values = [200,1.0,50,60,0.1,0.01,100]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
params = dict(penalty=penalty,C=c_values,solver=solver)
params
```

## Grid search cv

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold

grid = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=StratifiedKFold(n_splits=50))

```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
acc = accuracy_score(y_test,y_pred)
acc
con = confusion_matrix(y_test,y_pred)
con

pre_recall =  classification_report(y_test,y_pred)
acc,con,pre_recall
```

# Randomizied Search CV

```python
from sklearn.model_selection import RandomizedSearchCV

model = LogisticRegression()

randcv = RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randcv.fit(X_train,y_train)
```

```python
y_pred = randcv.predict(X_test)
y_pred
```

```python
acc = accuracy_score(y_test,y_pred)
acc
con = confusion_matrix(y_test,y_pred)
con

pre_recall =  classification_report(y_test,y_pred)
acc,con,pre_recall
```

```python
randcv.best_estimator_ , randcv.best_params_
```

# logistic regeression OVR

```python
X,y= make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=3,random_state=15)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
model =  LogisticRegression(multi_class='ovr')

```

```python


model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
acc = accuracy_score(y_test,y_pred)
acc
con = confusion_matrix(y_test,y_pred)
con

pre_recall =  classification_report(y_test,y_pred)
acc,con,
print(acc)
print(con)
print(pre_recall)
```
